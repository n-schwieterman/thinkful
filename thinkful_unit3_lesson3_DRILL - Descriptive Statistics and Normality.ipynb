```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
var1 = np.random.binomial(20, 0.5, 100)
var2 = np.random.gamma(20, 0.8,100)
var3 = np.random.pareto(100, 100)
var4 = np.random.power(5, 100)
var5 = np.random.geometric(0.6, 100)
var6 = np.random.poisson(8, 100)
```

```python
plt.hist(var1)
plt.axvline(var1.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(var1.mean()-var1.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(var1.mean()+var1.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
#Binomial distribution
```

The mean and standard deviation do provide useful info of the distribution

```python
norm1 = np.random.normal(5, 0.5, 100)
norm2 = np.random.normal(10, 1, 100)
norm3 = norm1 + norm2

plt.axvline(norm3.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(norm3.mean()+norm3.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(norm3.mean()-norm3.std(), color='r', linestyle='dashed', linewidth=2)

plt.hist(norm3)
plt.show()
```
